```python
def age_bins(x):
  if (x<25): return '1. < 25'
  if (x<35): return '2. 25-35'
  if (x<50): return '3. 36-50'
  if (x<65): return '4. 50-65'
  return '5. >65'

data.groupby(data.AGE.apply(lambda x: age_bins(x))).size()/ data.shape[0]
```
